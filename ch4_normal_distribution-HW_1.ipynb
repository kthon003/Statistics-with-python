The goal of this assignment is to understand how the normal probability plots (Q-Q plots) can be used to assess normality and visualize the skewness.

```python
import numpy as np
import pandas as pd

bdims = pd.read_csv('../datasets/bdims.csv')
bdims.head()
```

Create a subset `fdims`, which contains only the data for women.

```python
# start your code here:
fdims = bdims[bdims['sex'] == 0]

# end of your code.
```

Make a histogram and a normal probability plot for female biiliac (pelvic) diameter (`bii.di`), female elbow diameter (`elb.di`), female (`age`) and female chest depth (`che.de`).

We first to standardize the data by subtracting the mean, then dividing by the standard deviation. For example, you can create a variable called `bii` to store the data `fdims['bii.di']`, then assign `(bii - bii.mean()) / bii.std()` to a new variable called `bii_sd`. Do the same to other variables too.

```python
# start your code here:
fdims = fdims.assign(bii_sd = (fdims['bii.di'] - fdims['bii.di'].mean()) / fdims['bii.di'].std())
fdims = fdims.assign(elb_sd = (fdims['elb.di'] - fdims['elb.di'].mean()) / fdims['elb.di'].std())
fdims = fdims.assign(age_sd = (fdims['age'] - fdims['age'].mean()) / fdims['age'].std())
fdims = fdims.assign(che_sd = (fdims['che.de'] - fdims['che.de'].mean()) / fdims['che.de'].std())
fdims.head()
# end of your code.
```

We now create the histograms (with 15 bins) and normal probability plots. Please use `plt.subplot` to create 4 rows 2 columns plots, with each row for one variable.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 15)
plt.tight_layout(pad = 1) 
plt.subplots_adjust(hspace = 0.6, wspace = 0.6) 

from scipy.stats import probplot

# start your code here:

# plots for female bii.di
plt.subplot(3, 3, 1)
plt.hist(fdims['bii_sd'], color = 'gold', edgecolor = 'black', linewidth = 1.2, bins = 15)
plt.xlabel('bii_sd')
plt.ylabel('Counts')

plt.subplot(3, 3, 2)
probplot(fdims['bii_sd'], plot = plt)
plt.title('Normal bii_sd Plot (Data)')


probplot(fdims['bii_sd'], plot = plt)

# plots for female elb.di
plt.subplot(3, 3, 3)
plt.hist(fdims['elb_sd'], color = 'red', edgecolor = 'black', linewidth = 1.2, bins = 15)
plt.xlabel('elb_sd')
plt.ylabel('Counts')

plt.subplot(3, 3, 4)
probplot(fdims['elb_sd'], plot = plt)
plt.title('Normal elb_sd Plot (Data)')

probplot(fdims['elb_sd'], plot = plt)
# plots for female age
plt.subplot(3, 3, 5)
plt.hist(fdims['age_sd'], color = 'blue', edgecolor = 'black', linewidth = 1.2, bins = 15)
plt.xlabel('age_sd')
plt.ylabel('Counts')

plt.subplot(3, 3, 6)
probplot(fdims['age_sd'], plot = plt)
plt.title('Normal age_sd Plot (Data)')

probplot(fdims['age_sd'], plot = plt)
# plots for female che.de
plt.subplot(3, 3, 7)
plt.hist(fdims['che_sd'], color = 'green', edgecolor = 'black', linewidth = 1.2, bins = 15)
plt.xlabel('che_sd')
plt.ylabel('Counts')

plt.subplot(3, 3, 8)
probplot(fdims['che_sd'], plot = plt)
plt.title('Normal che_sd Plot (Data)')


probplot(fdims['che_sd'], plot = plt)
# end of your code
plt.show()
```

**Questions**: Based on the normal probability plots, how can you tell the skewness of the distribution? For example, if the normal probability plot is convex up (or concave down), what about the skewness? 

**Answer**: The skewness of bii.de is skewed to the right.
